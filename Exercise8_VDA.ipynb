Matplotlib Pie Chart - you can use pie() to draw pie chart

```python
g = np.array([35, 25, 25,15])
grader = ["Grader 1 = 35%", "Grader 2 = 25%", "Grader 3 = 25%", "Grader 4 = 15%",]
plt.pie(g, labels = grader)
plt.legend(labels = "Grader")
plt.legend(title = "Grader", labels = grader, loc = 'upper left')
plt.show()
```

Question 1: What is the correlation between the fire rate and wall penetration?



```python
import pandas as pd
import matplotlib.pyplot as plt

# Map wall penetration to numeric values
penetration_mapping = {"Low": 1, "Medium": 2, "High": 3}
df['Wall Penetration'] = df['Wall Penetration'].map(penetration_mapping)

# Prepare data for the box plot
data_to_plot = [df['Fire Rate'][df['Wall Penetration'] == i] for i in range(1, 4)]

# Create box plot
plt.figure(figsize=(10, 6))
plt.boxplot(data_to_plot, labels=['Low', 'Medium', 'High'])

# Adding title and labels
plt.title('Fire Rate Distribution by Wall Penetration Level')
plt.xlabel('Wall Penetration Level')
plt.ylabel('Fire Rate')
plt.grid()

# Show the plot
plt.show()
```
