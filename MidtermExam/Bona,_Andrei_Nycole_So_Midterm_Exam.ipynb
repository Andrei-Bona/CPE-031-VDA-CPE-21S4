```python
import pandas as pd
data = pd.read_csv('heart_2022_no_nans.csv')
```

```python
df = pd.DataFrame(data)
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
df.head(10)
```

```python
#1 What are the states included in this data set and which of them has the most and least number of people?
state_counts = df['State'].value_counts()

plt.figure(figsize=(15, 6))
state_counts.plot(kind='bar')
plt.title('Number of Occurrences per State')
plt.xlabel('State')
plt.ylabel('Count (Number of People)')
plt.show()

# Based on the graph, Washington has the most count while Virgin Islands has the least count. This graph indicates that the data set mostly consist of people living from Washington, but it also has a decent
# amount of data coming from different states.
```

```python
#2 Is there a difference between Male and Female when it comes to drinking alcohol?
sex_alcohol = df.groupby('Sex')['AlcoholDrinkers'].value_counts().unstack()

sex_alcohol.plot(kind='bar', figsize=(8, 6))
plt.title('Relationship between Sex and Alcohol Drinkers')
plt.xlabel('Sex')
plt.ylabel('Number of People')
plt.xticks(rotation=0)
plt.legend(title='Alcohol Drinkers')
plt.show()
# The bar graph shows that the number of Alcohol drinkers in Male is more than the Female but the number of Non-Alcohol Drinker in Female is more than the Male. This indicates that there is significant difference
# between Alcohol and Non-Alcohol drinkers in Male and Female
```
