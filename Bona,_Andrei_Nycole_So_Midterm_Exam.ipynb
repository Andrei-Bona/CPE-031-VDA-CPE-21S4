```python
import pandas as pd
data = pd.read_csv('heart_2022_no_nans.csv')
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#1.1 What are the states included in this data set and which of them has the most and least number of people?
state_counts = df['State'].value_counts()

plt.figure(figsize=(15, 6))
state_counts.plot(kind='bar')
plt.title('Number of Occurrences per State')
plt.xlabel('State')
plt.ylabel('Count (Number of People)')
plt.show()
'''
Based on the graph, Washington has the most count while Virgin Islands has
the least count. This graph indicates that the data set mostly consist of
people living from Washington, but it also has a decent
amount of data coming from different states.
'''
```

```python
#1.2 Is there a difference between Male and Female when it comes to drinking alcohol?
sex_alcohol = df.groupby('Sex')['AlcoholDrinkers'].value_counts().unstack()

sex_alcohol.plot(kind='bar', figsize=(8, 6))
plt.title('Relationship between Sex and Alcohol Drinkers')
plt.xlabel('Sex')
plt.ylabel('Number of People')
plt.xticks(rotation=0)
plt.legend(title='Alcohol Drinkers')
plt.show()
'''
The bar graph shows that the number of Alcohol drinkers in Male is more than
the Female but the number of Non-Alcohol Drinker in Female is more than the Male.
This indicates that there is significant difference
between Alcohol and Non-Alcohol drinkers in Male and Female
'''
```

```python
import pandas as pd
data1 = pd.read_csv('heart_2022_with_nans.csv')
```

```python
df1 = pd.DataFrame(data1)
```

```python
df1
```

```python
# 2.1 Does the distribution of general health in the dataset suggest any prevalent health concerns among the surveyed population?
health_counts = df1['GeneralHealth'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(health_counts, labels=health_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of General Health')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()

'''
The pie chart shows the distribution of general health among the repondents.
The largest slices indicate the most frequent responses. The significant portion of
the population reports "Good" followed by "Very Good" general health, it suggests a
relatively healthy population group in the dataset. In addition, if "Poor" or "Fair"
health dominate,  it could suggest potential public health issues that needs further investigation.
Further analysis might consider linking general health with other factors, such as lifestyle choices,
age, or access to healthcare, to gain deeper insights.
'''
```

```python
# 2.2 Does the respondents' sleep hour distribution indicate any common sleep patterns or possible sleep deficit issues?
plt.figure(figsize=(10, 6))
plt.hist(df1['SleepHours'], bins=range(0, 25), edgecolor='black')
plt.title('Distribution of Sleep Hours')
plt.xlabel('Hours of Sleep')
plt.ylabel('Number of Individuals')
plt.xticks(range(0,25))
plt.show()
'''
The histogram shows the distribution of sleep hours among the respondents.
Most people reported sleeping between 7 and 9 hours, with 8 hours being the most common.
The majority seem to sleep between 6 and 10 hours, while very few sleep less than 4 hours
or more than 11. This suggests that typical sleep duration for most individuals falls
between 7 and 9 hours, with fewer people.
'''
```
