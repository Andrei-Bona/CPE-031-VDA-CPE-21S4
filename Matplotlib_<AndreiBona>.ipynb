```python
import pandas as pd
data = pd.read_excel('Technological-Products-Sample-Data.xlsx')
```

```python
data.to_csv('Technological-Products-Sample-Data.csv')
df = pd.DataFrame(data)
```

```python
# clean the data set
df.drop(0, axis=0, inplace=True)
df.drop(1, axis=0, inplace=True)
df.drop(2, axis=0, inplace=True)
df.drop('Unnamed: 0', axis=1, inplace=True)
df.columns = ['Brand', 'Device', 'Model', 'Country of Origin', 'Date of Release', 'Price (USD)']
df.reset_index(drop=True, inplace=True)
```

```python
df.drop(0, axis=0, inplace=True)
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
# convert the datetime format
df['Date of Release'] = pd.to_datetime(df['Date of Release'])
df['Date of Release'] = df['Date of Release'].dt.strftime('%m/%d/%Y')
```

```python
df
```

```python
import matplotlib.pyplot as plt
```

```python
#a and b
plt.figure(figsize=(15, 5))
plt.plot(df['Date of Release'], df['Price (USD)'], 'o:r', linestyle = 'dotted', linewidth = 3, label = 'Price')
plt.legend(loc='lower right')
plt.title("Year and Price")
plt.xlabel("Year")
plt.ylabel("Price")
plt.xticks(rotation = 90)
plt.show()
```
